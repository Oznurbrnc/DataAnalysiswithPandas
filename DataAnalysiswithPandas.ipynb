
#### Additional Dataframe Operations

16. Perform a `merge` using "client_list.csv" and "customer_id_list.csv". Assign the resulting dataframe as `clients`.
17. Perform a `merge` using `clients` and "purchase_log.csv" and limit the subset to only clients who made purchases. Assign the resulting dataframe as `detailed_sales`.
18. Use `groupby` to find the client who spent the most money on purchases. Determine how much he/she spent. HINT: save the intermediate dataframe from using `groupby` as `spenders` before applying slicing to determine the client who spent the most money on purchases.
19. (BONUS) Modify the answer to problem 3 slightly to determine exactly what items where purchased by the top spending client.

```python
# CEVAP: Perform a merge using "client_list.csv" and "customer_id_list.csv". Assign the resulting dataframe as clients.
df5 = pd.read_csv("customer_id_list.csv")
clients = pd.merge(df5, df)
clients.head()

# ortak sütunlar: FIRST_NAME   LAST_NAME   GENDER   AGE
# farklı sütunlar : HAIR_COLOR   EYE_COLOR   CUSTOMER_ID
```

```python
# CEVAP: Perform a merge using clients and "purchase_log.csv" and limit the subset to only clients who made purchases. 
# Assign the resulting dataframe as detailed_sales.
dfpl = pd.read_csv("purchase_log.csv") 
detailed_sales = pd.merge(clients, dfpl, on='CUSTOMER_ID',how="inner") 
detailed_sales[['ITEM_DESCRIPTION' ,'PRICE']]

# price sorduğu için liste olarak detailed_sales[['ITEM_DESCRIPTION' ,'PRICE']] yazıyoruz. yoksa detailed_sales diye yazardık 
# clients içerisindeki sütunlar: CUSTOMER_ID	FIRST_NAME	LAST_NAME	GENDER	AGE	HAIR_COLOR	EYE_COLOR
# df_purchase içerisindeki sütunlar: CUSTOMER_ID	DATE	SKU	ITEM_DESCRIPTION	PRICE
```

```python
# CEVAP: Use groupby to find the client who spent the most money on purchases. 
# Determine how much he/she spent. HINT: save the intermediate dataframe from using groupby as spenders before applying 
# slicing to determine the client who spent the most money on purchases.
spenders = detailed_sales.groupby(["CUSTOMER_ID","FIRST_NAME", "LAST_NAME"])[["PRICE"]].sum()
spenders
```

```python
most_spender = spenders.sort_values(by="PRICE",ascending = False).iloc[0]
most_spender
```

```python
# CEVAP: (BONUS) Modify the answer to problem 3 slightly to determine exactly 
# what items where purchased by the top spending client.

spendersID = detailed_sales.groupby(['CUSTOMER_ID'])[['PRICE']].sum() 
spendersID
```

```python
detailed_sales[detailed_sales["CUSTOMER_ID"]==spendersID["PRICE"].idxmax()]["ITEM_DESCRIPTION"].to_frame()
```
